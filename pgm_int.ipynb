```python
def read_robbins_data(detections,i):
    dict={}
    dict["lat"]=float(detections.iloc[[i]].LAT_CIRC_IMG)
    dict["long"]=float(detections.iloc[[i]].LON_CIRC_IMG)
    dict["radius"]=float(detections.iloc[[i]].DIAM_CIRC_IMG)/2
    return dict
```

```python
def detect_craters(image):
    
    range=(int(image[-13:-11]),int(image[-13:-11])+30,int(image[-7:-4]),int(image[-7:-4])+60)
    detections=df_reset.loc[(range[0]<df_reset['LAT_CIRC_IMG']) & (df_reset['LAT_CIRC_IMG']<range[1]) & (range[2]<df_reset['LON_CIRC_IMG']) & (df_reset['LON_CIRC_IMG']<range[3])]
    rows=[]
    i=0
    while i<detections.shape[0]:
        row=read_robbins_data(detections,i)
        rows.append(row)
        i+=1
    return rows

detect_craters(filepath)
```

```python
def split_image(image):
    
    detections=detect_craters(image)
            
    for i in range(5159,5160):
            row=detections[i]
            y=row["lat"]
            x=row["long"]
            r=row["radius"]
            
            cut_img(image,x,y,r)
            
split_image(filepath)
```
